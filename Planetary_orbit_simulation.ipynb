```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gravitational constant
G = 1

# Mass of the star
M = 100

# Time step
dt = 0.001

# Number of simulation steps
steps = 100000
```

```python
# Initial position of planet
x = 1
y = 0
```

```python
# Distance from star
r0 = np.sqrt(x**2 + y**2)
```

```python
# Circular orbit velocity
v_circular = np.sqrt(G * M / r0)
```

Choose velocity
For circular orbit → v_circular
For elliptical orbit → 0.8 * v_circular

```python
vx = 0
vy = 0.8 * v_circular
```

```python
# Store orbit points
x_list = [x]
y_list = [y]
```

```python
# Velocity Verlet integration
for i in range(steps):

    r = np.sqrt(x**2 + y**2)

    ax = -G * M * x / (r**3)
    ay = -G * M * y / (r**3)

    # Update position
    x += vx * dt + 0.5 * ax * dt**2
    y += vy * dt + 0.5 * ay * dt**2

    # New acceleration
    r_new = np.sqrt(x**2 + y**2)
    ax_new = -G * M * x / (r_new**3)
    ay_new = -G * M * y / (r_new**3)

    # Update velocity
    vx += 0.5 * (ax + ax_new) * dt
    vy += 0.5 * (ay + ay_new) * dt

    x_list.append(x)
    y_list.append(y)
```

```python
# Plotting the graph
plt.figure(figsize=(7,7))

plt.plot(x_list, y_list, label="Planet Orbit")
plt.scatter(0, 0, color="red", label="Star")

plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("Planetary Orbit Simulation")
```

```python
# Ensure orbit centered around star
plt.axis('equal')
```

```python
# Move legend outside plot
plt.legend(loc="upper left", bbox_to_anchor=(1,1))

plt.grid()

plt.savefig("orbit_simulation.png", dpi=300, bbox_inches="tight")

plt.show()
```
